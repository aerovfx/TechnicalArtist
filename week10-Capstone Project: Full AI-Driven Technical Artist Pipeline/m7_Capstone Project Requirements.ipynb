### Capstone Project Requirements
🎯 “Build and Showcase an End-to-End AI Pipeline for Technical Art”
Each student must deliver a working pipeline that demonstrates:
Dataset: procedural dataset stored in Blob storage.
Model: trained ML model (e.g., clustering/regression/classification).
Deployment: FastAPI service exposing the model.
Integration:
Houdini HDA that queries predictions for sims/geometry.
Unity scene that consumes same model for gameplay or visualization.
Automation: simple reproducible pipeline (train → deploy → query).
Showcase: short video or presentation of the pipeline running.

Deliverables
GitHub repo (code + dataset samples + instructions).
Houdini HDA file + Unity package.
Recorded demo reel (2–3 minutes) showing the pipeline in action.
Technical write-up (documentation + lessons learned).

👉 By the end of Week 10, students will:
Have a portfolio-ready AI pipeline project.
Demonstrate integration of ML models into Houdini & Unity workflows.
Understand how to structure AI for production in game/VFX pipelines.
Be prepared to expand into studio-scale AI workflows.





Capstone Project Ideas
1. Procedural Asset Classifier & Tagging System
🎯 Mục tiêu: Xây dựng một AI pipeline để phân loại và gán tag cho tài sản procedural (geometry, textures, simulations).
Dataset:
Houdini-generated assets (rocks, trees, props, particles).
Procedural textures/materials.
Model:
Classification (CNN hoặc Random Forest cho metadata).
Predict asset category: “Rock / Plant / Debris / Building / Character Part”.
Integration:
Houdini HDA → Auto-tag nodes/geometry.
Unity → Auto-assign tags/layers for prefabs.
Use Case for TA:
Quản lý large procedural libraries.
Chuẩn bị dữ liệu cho pipeline LOD, optimization, auto-assembly.

2. AI-Driven Crowd Simulation & Behavior Clustering
🎯 Mục tiêu: Dùng AI để phân cụm và dự đoán hành vi cho đám đông trong game/film.
Dataset:
Houdini particle sims (agents, velocities, actions).
NPC motion capture data.
Model:
Clustering (K-means / DBSCAN) cho hành vi: Idle, Aggressive, Patrol, Flee.
Regression → Predict crowd density zones.
Integration:
Houdini: Crowd sim clustering → màu/nhóm agents trong viewport.
Unity: Import clustering → NPC AI states (e.g., aggressive cluster attacks player).
Use Case for TA:
Crowd VFX shots.
AI-driven gameplay systems trong Unity.

3. Smart Texture Manager & Quality Control
🎯 Mục tiêu: Dùng AI để quản lý, kiểm định chất lượng (QC) và tự động gợi ý textures/materials.
Dataset:
Texture library (procedural textures from Substance/Houdini + real scans).
Metadata (resolution, tiling, usage).
Model:
SVM/Decision Tree → QC textures (valid/invalid, seamless/non-seamless).
CNN → Predict material category (wood, stone, metal, fabric).
Integration:
Houdini: Auto-suggest texture/material for procedural assets.
Unity: Auto-assign correct texture sets for prefabs (PBR pipeline).
Use Case for TA:
Chuẩn hóa asset pipeline.
Giảm lỗi render/game build do asset không đạt chuẩn.

👉 Học viên có thể chọn một trong ba, hoặc thậm chí kết hợp để làm Capstone:
VD: Crowd Simulation + Smart Texture Manager → pipeline VFX/game AI + lookdev QC.




Capstone Project – Grading Criteria (Rubric)
1. Technical Implementation (40%)
40–36 (Excellent):
Pipeline hoàn chỉnh, chạy ổn định.
Mô hình AI được train/fine-tune với dữ liệu thực tế.
FastAPI deployment + Houdini/Unity integration hoạt động mượt.
35–28 (Good):
Pipeline chạy được, nhưng còn lỗi nhỏ.
Mô hình AI train được nhưng chưa tối ưu.
Integration có nhưng hạn chế (chỉ Houdini hoặc Unity).
27–20 (Fair):
Pipeline chỉ chạy từng phần.
Dùng pre-trained models, không fine-tune.
Integration còn sơ sài (demo script thay vì HDA/Unity prefab).
<20 (Poor):
Project không chạy được.
Chỉ lý thuyết, không có pipeline thực tế.

2. Creativity & Innovation (20%)
20–18 (Excellent):
Ý tưởng độc đáo, có ứng dụng thực tế rõ ràng trong TA pipeline.
Kết hợp nhiều kỹ thuật (clustering + QC + deployment).
Giải quyết vấn đề mà TA thường gặp (QC, sim speed, asset tagging).
17–14 (Good):
Ý tưởng hay nhưng ứng dụng chưa rõ ràng.
Có cải tiến nhỏ từ ví dụ trong lớp.
13–10 (Fair):
Ý tưởng đơn giản, lặp lại bài tập trong lớp.
Ít yếu tố sáng tạo.
<10 (Poor):
Ý tưởng không rõ ràng, không có yếu tố sáng tạo.

3. Integration & Pipeline Cohesion (25%)
25–23 (Excellent):
Liên kết chặt chẽ giữa dataset → model → deployment → Houdini/Unity.
Pipeline có thể lặp lại (reproducible).
Blob storage được sử dụng cho data management.
22–18 (Good):
Có pipeline đầy đủ nhưng chưa hoàn toàn tự động.
Integration chỉ một chiều (e.g., từ Houdini gọi API, chưa ngược lại).
17–13 (Fair):
Pipeline rời rạc, nhiều bước thủ công.
Integration chỉ demo, không đưa vào node/game object.
<13 (Poor):
Không có pipeline rõ ràng, các phần rời rạc.

4. Presentation & Documentation (15%)
15–14 (Excellent):
Demo reel rõ ràng (2–3 phút), highlight pipeline và kết quả.
Repo GitHub đầy đủ: code, dataset mẫu, hướng dẫn setup.
Documentation giải thích kiến trúc + use case TA.
13–11 (Good):
Demo reel có nhưng chưa mạch lạc.
Repo có code nhưng thiếu hướng dẫn.
10–8 (Fair):
Chỉ có demo ngắn hoặc video thô.
Documentation sơ sài.
<8 (Poor):
Không có demo hoặc tài liệu đi kèm.

Tổng điểm
90–100: Outstanding → Portfolio-ready project, thể hiện kỹ năng TA AI cao cấp.
80–89: Strong → Pipeline tốt, có thể show trong reel.
70–79: Pass → Có pipeline cơ bản, thiếu polish.

👉 Rubric này vừa đo kỹ thuật vừa đo sáng tạo và tính production. Nó sẽ giúp học viên biết mình phải làm gì để đạt project showcase chất lượng studio.


Capstone Project Report Outline
1. Abstract (200–300 words)
Giới thiệu ngắn gọn project: mục tiêu, vấn đề muốn giải quyết, và ý nghĩa trong pipeline của Technical Artist.
Tóm tắt kết quả đạt được (ví dụ: “AI clustering giúp giảm 40% thời gian setup crowd sim”).

2. Tools & Environment
Danh sách công cụ/phần mềm:
Python, Pytorch, FastAPI, Streamlit, Docker.
Houdini, Unity/Unreal (nếu có).
Cloud services (AWS S3, Azure Blob, GCP).
Hardware: GPU/CPU specs.
Dataset: nguồn gốc (procedural, mocap, texture scans, synthetic).

3. Methodology
Pipeline overview (diagram hoặc flowchart):
Dataset → Model Training → Deployment → Houdini/Unity Integration.
ML Models: mô tả thuật toán (CNN, clustering, regression, RL…).
Integration Steps: Houdini HDA, Unity prefab/script, FastAPI endpoints.
Challenges & Solutions: vấn đề gặp phải và cách xử lý.

4. Results & Evaluation
Quantitative Results (số liệu, biểu đồ):
Accuracy, clustering score, sim time prediction error…
Qualitative Results (visual):
Screenshots Houdini viewport, Unity demo, gif video.
Comparison: trước & sau khi áp dụng AI.

5. Demo & Pipeline Walkthrough
Link demo video (YouTube/Vimeo).
Hướng dẫn run project (tóm tắt README):
Clone repo → Run FastAPI → Call from Houdini Python Panel → See results.
Showcase end-to-end workflow.

6. Lessons Learned & Future Work
Những gì học viên rút ra (AI strengths, limitations, integration challenges).
Đề xuất cải thiện:
Sử dụng synthetic data.
Triển khai pipeline multi-user.
Thử nghiệm mô hình mới (Transformers, diffusion models).

7. References & Resources (optional)
Papers, tutorials, GitHub repos, course lectures.

👉 Format này giúp học viên có report dễ đọc, vừa academic vừa industry-ready. Nó cũng hỗ trợ khi nộp portfolio: recruiter hay supervisor sẽ thấy documentation + pipeline clarity.
